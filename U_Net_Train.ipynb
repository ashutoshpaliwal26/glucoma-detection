```python
IMG_SIZE = 256

images = []
od_img = []

for item in df["names"]:
    if not check_file(f"./data/cropped/full-fundus/{item}.png"):
        continue
    
    img = cv2.imread(f"./data/cropped/full-fundus/{item}.png")
    # g_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    r_img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))

    mask_img = cv2.imread(f"./data/cropped/optic-disc/{item}.png")
    # g_mask = cv2.cvtColor(mask_img, cv2.COLOR_BGR2GRAY)
    r_mask = cv2.resize(mask_img, (IMG_SIZE, IMG_SIZE))

    images.append(r_img)
    od_img.append(r_mask)
```

```python
print("Image size : ", len(images))
print("OD Images : ", len(od_img))
```

```python
images = np.array(images)
od_img = np.array(od_img)
```

```python
images = images.astype("float32") / 255.0
od_img = images.astype("float32") / 255.0
```

```python
e_dim_images = np.expand_dims(images, axis=-1)
e_dim_mask = np.expand_dims(od_img, axis=-1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(e_dim_images, e_dim_mask, test_size=0.1, random_state=42)
```

```python
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Conv2DTranspose, concatenate, BatchNormalization, Activation, Dropout
from tensorflow.keras.models import Model
```

```python
inputs = Input((IMG_SIZE, IMG_SIZE, 1))
```

```python
def conv_block(x, filters):
    x = Conv2D(filters, (3, 3), padding="same")(x)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)

    x = Conv2D(filters, (3, 3), padding="same")(x)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)
    return x
```

```python
# Encoder
c1 = conv_block(inputs, 64)
p1 = MaxPooling2D((2, 2))(c1)
c2 = conv_block(p1, 128)
p2 = MaxPooling2D((2, 2))(c2)
c3 = conv_block(p2, 256)
p3 = MaxPooling2D((2, 2))(c3)
c4 = conv_block(p3, 512)
p4 = MaxPooling2D((2, 2))(c4)
# Bottleneck
bn = conv_block(p4, 1024)
bn = Dropout(0.4)(bn)
# Decoder
u1 = Conv2DTranspose(512, (2, 2), strides=2, padding="same")(bn)
u1 = concatenate([u1, c4])
c5 = conv_block(u1, 512)
u2 = Conv2DTranspose(256, (2, 2), strides=2, padding="same")(c5)
u2 = concatenate([u2, c3])
c6 = conv_block(u2, 256)
u3 = Conv2DTranspose(128, (2, 2), strides=2, padding="same")(c6)
u3 = concatenate([u3, c2])
c7 = conv_block(u3, 128)
u4 = Conv2DTranspose(64, (2, 2), strides=2, padding="same")(c7)
u4 = concatenate([u4, c1])
c8 = conv_block(u4, 64)
# Output layer (binary mask)
outputs = Conv2D(1, (1, 1), activation="sigmoid")(c8)
model = Model(inputs, outputs)
```

```python
model.compile(loss="binary_crossentropy", optimizer="Adam", metrics=["accuracy"])
```
